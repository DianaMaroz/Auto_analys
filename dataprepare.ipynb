## Подготовка данных для проекта
### Добавление марки и модели в таблицу маркетинговые данные

```python
import pandas as pd
```

```python
# импортируем таблицы
df_market_data = pd.read_csv('Премиум авто - Маркетинговые данные.csv')
df_market_data.head(2)
```

```python
# Посмотрим уникальные значения столбца Domain
domain_uniq = df_market_data.Domain.unique()
domain_uniq
```

```python
# Посмотрим уникальные значения столбца Sourse
source_uniq = df_market_data.Source.unique()
source_uniq
```

```python
# Добавим таблицу Справочник и вычленим из нее модели
df_handbook = pd.read_csv('Премиум авто - Справочник.csv')
df_handbook.head()
```

```python
models = df_handbook['Модель'] 
models
```

```python
# Функция для получения модели из Goal Completion Location
def get_model(goal_completion_location):
    for m in models:
        if m in goal_completion_location:
            return m
        
```

```python
# Добавим колонку с моделью
df_market_data['model']= df_market_data['Goal Completion Location'].apply(get_model)
```

```python
df_market_data.head(10)
```

```python
df_handbook['Марка']= df_handbook['Марка'].str.replace(' ', '')
df_handbook['Марка']= df_handbook['Марка'].str.lower()
def rename_mark(mark):
    if mark == 'mercedes':
        return 'Mercedes'
    else:
        return 'BMW'
df_handbook['Марка']= df_handbook['Марка'].apply(rename_mark)
df_handbook['Марка']
```

```python
df_handbook.rename(columns = {'Модель':'model'}, inplace = True)
df_handbook.head(2)                             
```

```python
# Получим марки : если нет марки попробуем достать ее из Goal Completion Location
def get_brand(goal_compr_loc):
    if 'bmw' in goal_compr_loc.lower():
        return 'BMW'
    elif 'mercedes' in goal_compr_loc.lower():
        return "Mercedes"
    elif 'mersedes' in goal_compr_loc.lower():
        return "Mercedes"
    else:
        return 'Not set'
```

```python
df_market_data['mark2']= df_market_data['Goal Completion Location'].apply(get_brand)
df_market_data.head(20)
```

```python
df_new_market_data= pd.merge(df_market_data, df_handbook, on='model', how='left')
df_new_market_data.rename(columns = {'Марка':'brand'}, inplace = True)
#df_new_market_data.drop(['Цена'], axis=1, inplace = True)
df_new_market_data.head(20)
```

```python
#сохраним данные как новую таблицу
df_new_market_data.to_csv('new_market_data.csv', index=False)
```

```python
# получим таблицу с цепочками касаний для клиентов
df_CRM_data = pd.read_csv('Премиум авто - Данные из CRM.csv')
df_CRM_data.head()
```

```python
merged_data = pd.merge(df_new_market_data, df_CRM_data, on='Client ID', how='left')
merged_data.head()
```

```python
chain = merged_data.groupby(['Client ID']).apply(lambda x: x.astype(str).apply('->'.join))

chain
```

```python
df_new_market_data.to_excel('данные_с_марками.xlsx', index=False)
```

```python
chain.columns
```

```python
def sum_column(col):
    col = col.split('->')
    res = 0
    for i in col:
        if i.isdigit():
            res += int(i)
    return res
```

```python
chain['сумма конверсий']= chain['Конверсия'].apply(sum_column)
chain['сумма просчета']= chain['Просчет стоимости модели'].apply(sum_column)
chain['сумма приходов']= chain['Приход к диллеру'].apply(sum_column)
chain['сумма продаж']= chain['Продажа'].apply(sum_column)
                             
chain.head(20)
```

```python
def count_touch(col):
    col = col.split('->')
    return len(col)
```

```python
chain['количество касаний']= chain['City'].apply(count_touch)

                             
chain.head(20)
```

```python
chain.to_csv('chain.csv', index=False)
```
